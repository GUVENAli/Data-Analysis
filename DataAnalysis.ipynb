```python
class plotHist: 
    # Private initial variables
    def __init__(self):
        import pandas as pd
        self.__dataFileName = "iris.csv" # default
        self.__header = 0
        self.__delimiter = None
        self.__dataLengthToShow = 5
        self.__data = pd.DataFrame()
        self.__featuresData = pd.DataFrame()
        self.__labels = pd.DataFrame()
        self.__numericFeatures = pd.DataFrame() # For histogram
        self.__figWidth = 15
        self.__figHeight = 10
        self.__hbins = 20
        
    
    def setDataName(self, name):
        if(isinstance(name, str)):
            self.__dataFileName = name
        else:
            print("The data name must be string!")
    
    def getDataName(self):
        return self.__dataFileName
    
    def setHeader(self, h):
        if(isinstance(h, int)):
            self.__header = h
        else:
            print("The header must be integer!")
    
    def getHeader(self):
        return self.__header
    
    def setDelimiter(self, d):
        if(isinstance(d, str)):
            self.__delimiter = d
        else:
            print("The delimiter must be string!")
    
    def getDelimiter(self):
        return self.__delimiter
    
    def setDataLengthToShow(self, l):
        if(isinstance(l, int)):
            self.__dataLengthToShow = l
        else:
            print("The data length to show must be integer!")
    
    def getDataLengthToShow(self):
        return self.__dataLengthToShow
    
    def setFeaturesData(self,f):
        self.__featuresData = f
        
    def getFeaturesData(self):
        return self.__featuresData
    
    def setLabel(self,y):
        self.__labels = y
        
    def getLabels(self):
        return self.__labels
    
    def setFigWidth(self, w):
        if(isinstance(w, int)):
            self.__figWidth = w
        else:
            print("The figure width must be integer!")
    
    def getFigWidth(self):
        return self.__figWidth
    
    def setFigHeight(self, h):
        if(isinstance(h, int)):
            self.__figHeight = h
        else:
            print("The figure height must be integer!")
    
    def getFigHeight(self):
        return self.__figHeight
    
    def setHBins(self, b):
        if(isinstance(b, int)):
            self.__hbins = b
        else:
            print("The histogram bin number must be integer!")
    
    def getHBins(self):
        return self.__hbins
    
    def readData(self):
        import pandas as pd
        self.__data = pd.read_csv(self.__dataFileName, header = self.__header, delimiter = self.__delimiter)
        
    def showData(self):
        #data = pd.read_csv(self._dataFileName, header = self__header, delimiter = self.__delimiter)
        return self.__data.head(self.__dataLengthToShow)
        
    def findNumericFeatures(self):
        names = []
        for i in self.__data.columns:
            if(isinstance(self.__data[i].sum(),(int,float,complex))):
                names.append(i)
        print("Found numeric features:",names)
        self.__numericFeatures = self.__data[names]
        
    def getNumericFeature(self):
        return self.__numericFeatures
    
    def plotHistogram(self):
        from matplotlib import pyplot as plt
        plt.figure(figsize=(self.__figWidth,self.__figHeight))
        for i in self.__numericFeatures:
            plt.hist(self.__numericFeatures[i], bins = self.__hbins)
        plt.legend((self.__numericFeatures.columns))
        plt.xlabel("Features")
        plt.ylabel("Count")
    
    def plotBoxPlot(self):
        from matplotlib import pyplot as plt
        plt.figure(figsize=(self.__figWidth,self.__figHeight))
        self.__numericFeatures.boxplot()        
```

```python
Data = plotHist() # Object created
```

```python
Data.setDataName = "iris.csv"
Data.readData()
Data.setDataLengthToShow(15)
Data.showData()
```

```python
Data.plotHistogram()
Data.plotBoxPlot()
```
